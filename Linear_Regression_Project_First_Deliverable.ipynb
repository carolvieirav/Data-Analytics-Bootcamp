```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```
diamonds = pd.read_csv('/content/diamonds.csv').reset_index(drop=True)
```

```
rick_diamonds = pd.read_csv('/content/rick_diamonds.csv').reset_index(drop=True)
```

```
diamonds.head()
```

```
rick_diamonds.head()
```

```
dmean = diamonds['price'].mean()
dmean
```

```
diamonds['price_predicted'] = dmean
diamonds.head()
```

```
rick_diamonds['price_predicted'] = dmean
rick_diamonds
```

```
from sklearn.metrics import mean_squared_error
```

```
rmse = np.sqrt(mean_squared_error(diamonds['price'], diamonds['price_predicted']))
rmse
```
